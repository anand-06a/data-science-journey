```python
# i like python
# vowels '*' 'aeiou' 
# * l*k* pyth*n
text='i like python'
vowels='aeiou'
for i in text:
    c=0
    for j in vowels:
        if i==j:
            print("*",end='')
            c=1
            break
    if c==0:
        print(i,end='')
    
```

```python
text=input()
vowels='aeiouAEIOU'
for i in text:
    c=0
    for j in vowels:
        if i==j:
            print("*",end='')
            c=1
            break
    if c==0:
        print(i,end='')
```

```python
s=input()
vowel='aeiouAEIOU'
res=''
for i in s:
    if i in vowel:
        res+='*'
    else:
        res+=i
print(res)
```
