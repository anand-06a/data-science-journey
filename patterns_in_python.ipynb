```python
rows=5
for i in range(1,rows +1):
    for j in range(rows-i):
        print("  ",end="")
    for k in range(2*i-1):
        print("*",end="")
    print()
```

```python
# star pyramid
rows=5
for i in range(1,rows +1):
    for j in range(rows-i):
        print("",end="")
    for k in range(2*i-1):
        print("0",end="")
    print()
```

```python
# number pyramid
rows=5
for i in range(1,rows+1):
    for j in range(rows-i):
        print("",end="")
    for k in range(i):
        print(i,end="")
    print()
```

```python
# alphabet pyramid
rows=5
for i in range(1,rows+1):
    for j in range(rows-i):
        print(" ",end="")
    for k in range(i):
        print(chr(64+i),end="")
    print()
```

```python
# inverted str pyramid
rows=5
for i in range(rows,0,-1):
    for j in range(rows-i):
        print(" ",end="")
    for k in range(2*i-1):
        print("*",end="")
    print()
```

```python
# full pyramid(1ton)
rows=5
for i in range(1,rows+1):
    for j in range(rows-i):
        print(" ",end=" ")
    for k in range(1,i+1):
        print(k,end=" ")
    print()
```
