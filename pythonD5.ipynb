```python
import random

print("==== NUMBER GUESSING GAME ====")

# GENERATE A RANDOM NUMBER  B/W 5 AND 10
secret_number = random.randint(5,10)

attempts = 0

while True:
    guess = int(input("Guess a number between 5 and 10:"))
    attempts += 1
    
    if guess == secret_number:
        print("Congratulations! You guessed the correct number.")
        print(f"You guessed it in {attempts} attempt(s).")
        break
        
    elif guess < secret_number:
        print("Too low! Try again.")
        
    else:
        print("Too high! Try again.")
```

```python
text = "i like python"
vowels = "aeiou"

for i in text:
    c = 0
    
    for j in vowels:
        if i == j:
            print("*", end="")
            c = 1
            break
    
    if c == 0:
        print(i, end="")
```
