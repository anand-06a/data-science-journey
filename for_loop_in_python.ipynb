```python
# for loops in python
# it is used to iterate over a sequence(like a list,tuple,string,or range) and execute a block of code 
#repeatedly for each element in the sequence.
# the basic structure of a for loop
#-- a for loop allows you to repeat a block of code a fixed number of times,or once for each element in a sequence
```

```python
# syntax
for item in sequence:
    # code to execute for each item in the sequence
```

```python
bag=['red','green','blue']
for ball in bag:
    print(ball,end=' ')
```

```python
for i in range(1,11,2):
    print(i)
```

```python
# looping over a string
name='charan'
for letter in name:
    print(letter)
```

```python
name='charan'
for letter in name:
    print(letter,end=' ')
```

```python
name='charan'
for letter in name:
    print(letter*3)
```

```python
name='charan'
for index, letter in enumerate(name):
    print(letter*(index+1))
```

```python
l=[11,34,56,876]
for index,num in enumerate(l):
    print(f"{num} is in {index}th index")
```

```python
for num in l:
    print(num)
    if num==56:
        break
else:
    print("all printed")
```

```python
d={"name":"charan","age":22,"income":30000}
for key,value in d.items():
    print(key,':',value)
```

```python
# nested for loops
# it means a loop inside another loop.this useful when working with multi-level data,like lists inside lists
```

```python
for i in range(1,11):
    print(f"2X{i}={2*i}")
```

```python
for i in range(2,11):
    for j in range(1,11):
        print(f"{i}X{j}={i*j}")
```

```python
student_marks={'anand':76,'bhunu':6,'siddhu':74}
for student,marks in student_marks.items():
    print(f"{student}:{marks}")
```

```python
students=['chandu','darshan','narendra']
marks=[25,78,56]
student_marks={}
for i in range(1,len(students),2):
    student_marks[students[i]]=marks[i]
print(student_marks)
```

```python
# list comprehension
# exp for item in collection 
l=[1,2,3,4,5]
dl=[item**2 for item in l]
print(dl)
```

```python
l=[1,2,3,4,5]
dl=[item**2 for item in l if item%2==0]
print(dl)
```

```python
l=[1,2,3,4,5]
edl=[item**2 for item in l if item%2==0]
print(edl)
```
